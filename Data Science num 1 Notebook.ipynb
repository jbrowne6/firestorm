## Interpreting the Output
That probably didn't mean much to you.

Iris is a famous dataset that many data scientists use to learn how to work with different tools. It is often referred to as a toy dataset because many people like to play with it.

Datasets come in many different forms. Typically, the dataset comes as a .csv file. In this case we are using the version provided by sklearn. sklearn (scikit-learn) is a very popular machine learning package that bundles popular toy datasets and popular analysis tools. 

Lets try and make sense of the output from above.
***

<br><br>

```python
iris.data
```

<br><br>
***
Above we have an array of lists. Arrays aren't data structures in Python, but they are in many other programming languages. There are two popular Python packages that implement arrays to make data science easier: pandas and numpy.

Back to the data. Above there are 150 lists in the array each containing 4 entries. Each of the 150 lists represents the measurements of a single flower. We call this an instance (synonyms: samples and records). The 4 entries in each list are called features (synonyms: dimensions and attributes). So we say each instance has 4 features. 

What are the features? Good question.
***

<br><br>

```python
iris.feature_names
```

<br><br>
***
You can look at this website to learn what these features actually mean. __[iris dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set)__

This dataset actually has one more feature that sklearn has decided to separate from the rest of the data. This special feature is called the label (or target).
***

<br><br>

```python
# These are the labels
print(iris.target)
print("\n\n")
# This is what the labels mean
print(iris.target_names)
```

<br>
<br>

***
There are 150 labels, 1 for each instance. If the label is a 0 then the instance is setosa. If the label is a 1, versicolor; 2, virginica.

Lets take a look at one of the instances.
***

<br>
<br>

```python
# Lets look at instance 135.
print(iris.data[135])
print("\n\n")
# Here is the first feature of instance 135
print(iris.data[135][0])
print("\n\n")
# Here is the label
print(iris.target[135])
print("\n\n")
# We can print the label better than that.
print(f"The label for instance 135 is {iris.target_names[iris.target[135]]}")
```

Lets try and look at the data graphically.
***

<br><br>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import matplotlib


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


def get_feature(feature_num, np_array):
    feature = []
    for instance in np_array:
        feature.append(instance[feature_num])
    return np.array(feature)

# Collect the setosa data
x_setosa = get_feature(0, iris.data[0:50])
y_setosa = get_feature(1, iris.data[0:50])
z_setosa = get_feature(2, iris.data[0:50])
ax.scatter(x_setosa, y_setosa, z_setosa, c='r', marker='o',)

# Collect the versicolor data
x_versicolor = get_feature(0, iris.data[50:100])
y_versicolor = get_feature(1, iris.data[50:100])
z_versicolor = get_feature(2, iris.data[50:100])
ax.scatter(x_versicolor, y_versicolor, z_versicolor, c='b', marker='o')

# Collect the virginica data
x_virginica = get_feature(0, iris.data[100:150])
y_virginica = get_feature(1, iris.data[100:150])
z_virginica = get_feature(2, iris.data[100:150])
ax.scatter(x_virginica, y_virginica, z_virginica, c='g', marker='o')

# Set the labels for the axes
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])

# Create the legend
scatter1_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", c='r', marker = 'o')
scatter2_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", c='b', marker = 'o')
scatter3_proxy = matplotlib.lines.Line2D([0],[0], linestyle="none", c='g', marker = 'o')
ax.legend([scatter1_proxy, scatter2_proxy, scatter3_proxy], ['setosa', 'versicolor', 'virginica'], numpoints = 1)

plt.show()

```
